```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    self.head=Node(a[0],None,None)
    tail=self.head
    for i in range (1,len(a)):
      nw_n=Node(a[i],None,None)
      nw_n.prev=tail
      tail.next=nw_n
      tail=nw_n
    tail.next=self.head
    self.head.prev=tail

  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    count = 0
    data=self.head
    if data != None:
      count+=1
      data=data.next
    while data != self.head:
      count+=1
      data=data.next
    return count
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    head=self.head.next
    print(self.head.element, end=' ')
    while head != self.head:
      print(head.element, end=' ')
      head=head.next
    print()
  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    head=self.head.prev
    while head != self.head:
      print(head.element, end=' ')
      head=head.prev
    print(self.head.element, end=' ')
    print()

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    if idx<0 or idx > self.countNode():
      return None
    else:
      count = 0
      data=self.head
      if data != None:
        if count == idx:
          return data
        data=data.next     
      while data != self.head:
        count+=1
        if count==idx:
          return data        
        data=data.next
      return data  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count=0
    data=self.head.next
    if elem==self.head.element:
      return count
    else:
      while data != self.head:
        count+=1
        data=data.next
      return count    

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    if idx<0 or idx>self.countNode():
      return 'Ivalid index'
    else:
      if idx==0:
        newnode = Node(elem,None,None)
        nodeat = self.nodeAt(idx)
        pre = nodeat.prev
        newnode.prev = pre
        newnode.next = nodeat
        nodeat.prev = newnode
        pre.next = newnode
        self.head = newnode
      elif idx == self.countNode():
        newnode = Node(elem,None,None)
        nodeat = self.nodeAt(idx-1)
        newnode.prev = nodeat
        newnode.next = self.head
        nodeat.next = newnode
        self.head.prev = newnode
      else:
        newnode = Node(elem,None,None)
        nodeat = self.nodeAt(idx)
        pre = nodeat.prev
        newnode.prev = pre
        newnode.next = nodeat
        nodeat.prev = newnode
        pre.next = newnode
    
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    if idx<0 or idx>self.countNode():
      return str(None)
    else:
      if idx == 0:
        newnode=self.nodeAt(idx)
        temp1=newnode.prev
        temp2=newnode.next
        temp2.prev=temp1
        temp1.next=temp2
        self.head=temp2
        return str(newnode.element)
      else:
        newnode=self.nodeAt(idx)
        temp1=newnode.prev
        temp2=newnode.next
        temp2.prev=temp1
        temp1.next=temp2
        return str(newnode.element)
```
